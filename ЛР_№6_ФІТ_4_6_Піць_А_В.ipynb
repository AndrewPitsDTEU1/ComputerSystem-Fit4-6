```python
!pip install -q ultralytics
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')  # або yolov8s.pt для кращої точності

model.train(
    data='/content/datasets/data.yaml',
    epochs=20,
    imgsz=640,
    batch=16
)
```

```python
# Запуск оцінки на валідаційній вибірці
model.val()

# Вивід графіка результатів навчання
from IPython.display import Image
Image(filename='runs/detect/train/results.png')
```

```python
# Детекція на тестовому наборі
model.predict(
    source='/content/datasets/test/images',
    conf=0.3,
    save=True
)

# Відображення перших 3 результатів
import os
from IPython.display import display, Image

for fname in os.listdir('runs/detect/predict')[:3]:
    display(Image(filename=os.path.join('runs/detect/predict', fname)))
```
