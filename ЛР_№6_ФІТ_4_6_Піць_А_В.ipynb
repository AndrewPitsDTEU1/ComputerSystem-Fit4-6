```python
!pip install -q ultralytics
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')  # –∞–±–æ yolov8s.pt –¥–ª—è –∫—Ä–∞—â–æ—ó —Ç–æ—á–Ω–æ—Å—Ç—ñ

model.train(
    data='/content/datasets/data.yaml',
    epochs=20,
    imgsz=640,
    batch=16
)
```

```python
# –ó–∞–ø—É—Å–∫ –æ—Ü—ñ–Ω–∫–∏ –Ω–∞ –≤–∞–ª—ñ–¥–∞—Ü—ñ–π–Ω—ñ–π –≤–∏–±—ñ—Ä—Ü—ñ
model.val()

# –í–∏–≤—ñ–¥ –≥—Ä–∞—Ñ—ñ–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤ –Ω–∞–≤—á–∞–Ω–Ω—è
from IPython.display import Image
Image(filename='runs/detect/train/results.png')
```

```python
# –î–µ—Ç–µ–∫—Ü—ñ—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º—É –Ω–∞–±–æ—Ä—ñ
model.predict(
    source='/content/datasets/test/images',
    conf=0.3,
    save=True
)

# –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –ø–µ—Ä—à–∏—Ö 3 —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
import os
from IPython.display import display, Image

for fname in os.listdir('runs/detect/predict')[:3]:
    display(Image(filename=os.path.join('runs/detect/predict', fname)))
```
